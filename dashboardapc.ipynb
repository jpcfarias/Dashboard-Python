```python
#separei a coluna genero em 3 pois existiam muitos generos na mesma celular e o algoritimo lia apenas o primeiro, selecionamos alguns dentre todos os generos
#importando o pandas e o plotly
import pandas as pd
import plotly.express as px
#lendo o exel e transformando as colunas em lista
#df2 = pd.read_excel('https://docs.google.com/spreadsheets/d/e/2PACX-1vRRCqV7ulwZPdT6YJIW9XVcanKGIauDLKKgAKBAcV2W1PBIMMcW4S7Ch2dXc2IygQ/pub?output=xlsx')
df2 = pd.read_csv('https://raw.githubusercontent.com/jpcfarias/dashboard-python/main/steam-_1_%201.xlsx%20-%20Worksheet.csv')
generos = df2['genres'].tolist()
generos1 = df2['genres1'].tolist()
generos2 = df2['genres2'].tolist()
#definindo as variaveis
acao = 0
indie = 0
estrategia = 0
rpg = 0
casual = 0
simulacao = 0
corrida = 0
aventura = 0
esporte = 0
#lendo a primeira coluna
for i in range(len(generos)):
  if generos[i] == 'Action':
    acao = acao + 1
  elif generos[i] == 'RPG':
    rpg = rpg + 1
  elif generos[i] == 'Indie':
    indie = indie + 1
  elif generos[i] == 'Strategy':
    estrategia = estrategia + 1
  elif generos[i] == 'Casual':
    casual = casual + 1
  elif generos[i] == 'Simulation':
    simulacao = simulacao + 1
  elif generos[i] == 'Racing':
    corrida = corrida + 1
  elif generos[i] == 'Adventure':
    aventura = aventura + 1
  elif generos[i] == 'Sports':
    esporte = esporte + 1
  
#lendo a segunda coluna
for i in range(len(generos)):
  if generos1[i] == 'Action':
    acao = acao + 1
  elif generos1[i] == 'RPG':
    rpg = rpg + 1
  elif generos1[i] == 'Indie':
    indie = indie + 1
  elif generos1[i] == 'Strategy':
    estrategia = estrategia + 1
  elif generos[i] == 'Casual':
    casual = casual + 1
  elif generos[i] == 'Simulation':
    simulacao = simulacao + 1
  elif generos[i] == 'Racing':
    corrida = corrida + 1
  elif generos[i] == 'Adventure':
    aventura = aventura + 1
  elif generos[i] == 'Sports':
    esporte = esporte + 1
  
#lendo a terceira
for i in range(len(generos)):
  if generos2[i] == 'Action':
    acao = acao + 1
  elif generos2[i] == 'RPG':
    rpg = rpg + 1
  elif generos2[i] == 'Indie':
    indie = indie + 1
  elif generos2[i] == 'Strategy':
    estrategia = estrategia + 1
  elif generos[i] == 'Casual':
    casual = casual + 1
  elif generos[i] == 'Simulation':
    simulacao = simulacao + 1
  elif generos[i] == 'Racing':
    corrida = corrida + 1
  elif generos[i] == 'Adventure':
    aventura = aventura + 1
  elif generos[i] == 'Sports':
    esporte = esporte + 1
  
#criando um novo dicionario
dataframegenero = {
    'Generos' : ['Açao', 'RPG', 'Indie','Estrategia', 'Casual', 'Simulacao', 'Corrida', 'Aventura', 'Esporte'],
    'Quantide de jogos' : [acao, rpg, indie, estrategia, casual, simulacao, corrida, aventura, esporte]
}
#printando o grafico
fig = px.pie(dataframegenero, values='Quantide de jogos', names= 'Generos', title='Quantidade de jogos por genero')
fig.show()
```
